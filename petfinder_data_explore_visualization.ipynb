```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("./petfinder_data/train/train.csv")
df
```

## Data Fields

- PetID - Unique hash ID of pet profile
- AdoptionSpeed - Categorical speed of adoption. Lower is faster. This is the value to predict. See below section for more info.
- Type - Type of animal (1 = Dog, 2 = Cat)
- Name - Name of pet (Empty if not named)
- Age - Age of pet when listed, in months
- Breed1 - Primary breed of pet (Refer to BreedLabels dictionary)
- Breed2 - Secondary breed of pet, if pet is of mixed breed (Refer to BreedLabels dictionary)
- Gender - Gender of pet (1 = Male, 2 = Female, 3 = Mixed, if profile represents group of pets)
- Color1 - Color 1 of pet (Refer to ColorLabels dictionary)
- Color2 - Color 2 of pet (Refer to ColorLabels dictionary)
- Color3 - Color 3 of pet (Refer to ColorLabels dictionary)
- MaturitySize - Size at maturity (1 = Small, 2 = Medium, 3 = Large, 4 = Extra Large, 0 = Not Specified)
- FurLength - Fur length (1 = Short, 2 = Medium, 3 = Long, 0 = Not Specified)
- Vaccinated - Pet has been vaccinated (1 = Yes, 2 = No, 3 = Not Sure)
- Dewormed - Pet has been dewormed (1 = Yes, 2 = No, 3 = Not Sure)
- Sterilized - Pet has been spayed / neutered (1 = Yes, 2 = No, 3 = Not Sure)
- Health - Health Condition (1 = Healthy, 2 = Minor Injury, 3 = Serious Injury, 0 = Not Specified)
- Quantity - Number of pets represented in profile
- Fee - Adoption fee (0 = Free)
- State - State location in Malaysia (Refer to StateLabels dictionary)
- RescuerID - Unique hash ID of rescuer
- VideoAmt - Total uploaded videos for this pet
- PhotoAmt - Total uploaded photos for this pet
- Description - Profile write-up for this pet. The primary language used is English, with some in Malay or Chinese.

```python

adoption_speed = pd.DataFrame(
    {
        "AdoptionSpeed": df["AdoptionSpeed"].value_counts().sort_index(ascending=False).index.values,
        "Amount": df["AdoptionSpeed"].value_counts()[:]
    }
)
adoption_speed
```

```python
fig = px.bar(adoption_speed, x="AdoptionSpeed", y="Amount", color="AdoptionSpeed", title="Amount of each Adoption Speed")
fig.show()
```

```python
df["Type"].replace(
    [1, 2],
    ["Dog", "Cat"],
    inplace=True
)
df
```

```python
px.histogram(df, x="Type", color="Type", title="Amount of cat and dog")
```

```python
px.histogram(
    df, x="AdoptionSpeed", 
    color="Type", 
    barmode="group",
    title="Dog And Cat with Adoption"
).show()
```

```python

df['Name'] = df['Name'].fillna('Unnamed')
df['Has_name'] = 1
df.loc[df['Name'] == 'Unnamed', 'Has_name'] = 0
```

```python
px.histogram(
    df, x="Has_name", 
    color="AdoptionSpeed", 
    barmode="group",
    title="Pet Has Name And No Name with Adoption Speed"
).show()
```

```python
px.histogram(df, x="Age", title="Amount of Pets in each age")
```

```python



fig = px.pie(df, values="AdoptionSpeed", names="Age", title="Amount of AdoptionSpeed on Age", height=1000, color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_traces(textposition='inside', textinfo='percent+label').show()
```

```python

fig = px.histogram(df, x="AdoptionSpeed", color="Vaccinated", barmode="group", title=" Relationship between Vaccinated and Adoption Speed ")
fig.show()
```

```python
df_cp = df[["Fee", "AdoptionSpeed"]]
df_cp["Fee"].replace([0], ["Free"], inplace=True)

fig = px.pie(df_cp, names="Fee", title="AdoptionSpeed And Fee", height=1000, color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_traces(textposition='inside', textinfo='percent+label').show()
```

```python
filter = 10
result = df[df['Fee'] < filter]
px.histogram(result, x="Fee", color="AdoptionSpeed", barmode="group", title=f"Adopion Speed on Price from 0 -> {filter}")
```
