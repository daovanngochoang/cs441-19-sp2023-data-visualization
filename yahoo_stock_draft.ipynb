```python
class StockReader:
    def __init__(self, company_symbols: list, start_date: str = None, end_date: str = None) -> None:
        print("CONNECT AND FETCH DATA ...")
        self.__stock_price_data = self.__read_data(
            company_symbols, start_date, end_date)
            
        self.__current_symbols = company_symbols
        self.__start_date = start_date
        self.__end_date = end_date
        self.info = {}
        self.__get_info()
        self.stock_price = self.__stock_price_data
        
        
    def __get_info(self):
        for symbol in self.__current_symbols:
            print(f'fetching data of {symbol}')
            self.info[symbol] = yf.Ticker(symbol).info
    

    def __read_data(self, company_symbols, start_date=None, end_date=None):
        start = start_date if start_date != None else "2006-01-01"
        end = end_date if end_date != None else datetime.now().strftime('%Y-%m-%d')

        tickers = yf.Tickers(company_symbols)

        tickers_hist = tickers.history(start=start, end=end)

        return tickers_hist

    def __draw(self, df):
        px.line(df.reset_index(), y=companies, x="Date").show()

    def draw_close_price(self):
        self.__draw(self.__stock_price_data.Close)

    def draw_volume(self):
        self.__draw(self.__stock_price_data.Volume)

    def draw_high_price(self):
        self.__draw(self.__stock_price_data.High)

    def draw_low_price(self):
        self.__draw(self.__stock_price_data.Low)

    def draw_price_fluctuate(self, start_date, end_date):
        pass
```

```python
import plotly.graph_objects as go

# create first trace
trace1 = go.Scatter(
    x=[1, 2, 3],
    y=[4, 5, 6],
    name='Trace 1'
)

# create second trace
trace2 = go.Scatter(
    x=[1, 2, 3],
    y=[10, 11, 12],
    name='Trace 2',
    yaxis='y2'
)

# create layout with 2 y-axes
layout = go.Layout(
    title='Two Traces with Different Y-Axes',
    yaxis=dict(
        title='Trace 1'
    ),
    yaxis2=dict(
        title='Trace 2',
        overlaying='y',
        side='right'
    )
)

# create figure with both traces and layout
fig = go.Figure(data=[trace1, trace2], layout=layout)

# show figure
fig.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

ticker = "MSFT"
start_date = "2019-01-01"
end_date = "2023-01-01"

data = yf.download(ticker, start=start_date, end=end_date)
prices = data["Adj Close"]

plt.plot(prices)
plt.title(f"{ticker} Price Change from {start_date} to {end_date}")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import yfinance as yf

# Get MSFT data from Yahoo Finance
msft = yf.Ticker("MSFT")
data = msft.history(start="2019-01-01", end="2023-01-01")

# Create a figure with two subplots
fig = make_subplots(rows=2, cols=0)

# Add the first subplot
fig.add_trace(
    go.Scatter(x=data.index, y=data["Close"], name="MSFT Close"),
    row=1,
    col=0
)

# Add the second subplot
fig.add_trace(
    go.Scatter(x=data.index, y=data["Volume"], name="MSFT Volume"),
    row=2,
    col=0
)

# Update the layout of the figure
fig.update_layout(title="MSFT Stock Data", height=500, width=1000)

# Show the figure
fig.show()
```

```python
import plotly.graph_objs as go
import yfinance as yf

# Get Microsoft stock data from Yahoo Finance
msft = yf.Ticker("MSFT")
data = msft.history(period="5y")

# Create trace for the closing price data
trace1 = go.Scatter(x=data.index, y=data['Close'], name="MSFT Close")

# Create trace for the volume data
trace2 = go.Bar(x=data.index, y=data['Volume'], name="MSFT Volume")

# Create layout for the figure
layout = go.Layout(title="Microsoft Stock Data",
                   xaxis=dict(title="Date"),
                   yaxis=dict(title="Price"),
                   plot_bgcolor='black',
                   paper_bgcolor='black')

# Create figure object with the data and layout
fig = go.Figure(data=[trace1, trace2], layout=layout)

# Show the figure
fig.show()
```

```python
import yfinance as yf
import plotly.graph_objs as go

# Get MSFT data from 2019 to 2023
msft = yf.Ticker("MSFT")
msft_data = msft.history(start="2019-01-01", end="2023-01-01")

# Create a trace for MSFT close price
msft_trace = go.Line(x=msft_data.index, y=msft_data["Close"], name="MSFT")

# Create a layout with a black background
layout = go.Layout(
    paper_bgcolor="black",
    plot_bgcolor="black",
    font=dict(color="white")
)

# Create a figure with the MSFT trace and layout
fig = go.Figure(msft_trace, layout=layout)

# Display the figure
fig.show()
```
